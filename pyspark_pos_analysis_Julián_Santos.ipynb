```python

import nltk 
import string
import collections
from pyspark import SparkContext
from nltk import pos_tag
from nltk import RegexpParser
from nltk.tokenize import word_tokenize
from nltk.corpus import brown
```

```python
rdd_reuters = sc.textFile("reuters.txt")
type(u'lama')
string.punctuation
punc = '!"#$%&\'()*+,./:;<=>?@[\\]^_`{|}~“'
```

```python
def uni_to_clean_str(x):
    converted = x
    lowercased_str = converted.lower()
    lowercased_str = lowercased_str.replace(",","")
    lowercased_str = lowercased_str.replace("!","")
    lowercased_str = lowercased_str.replace(".","")
    lowercased_str = lowercased_str.replace("…","")
    lowercased_str = lowercased_str.replace(";","")
    # for more difficult cases use re.split(' A|B')
    lowercased_str = lowercased_str.replace('--',' ')
    lowercased_str = lowercased_str.replace('-',' ')
    lowercased_str = lowercased_str.replace('?',' ')
    lowercased_str = lowercased_str.replace('(',' ')
    lowercased_str = lowercased_str.replace(')',' ')
    lowercased_str = lowercased_str.replace('"',' ')
    lowercased_str = lowercased_str.replace("'",' ')
    lowercased_str = lowercased_str.replace("$",' ')
    clean_str = lowercased_str.translate(punc) #
    return clean_str
```

```python
one_RDD = rdd_reuters.flatMap(lambda x: uni_to_clean_str(x).split())
one_RDD = one_RDD.map(lambda x: (x,1))
one_RDD = one_RDD.reduceByKey(lambda x,y: x + y)
#one_RDD.take(15)
one_RDD.collect()
```

```python
file_content = open("reuters.txt").read()
tokens = word_tokenize(uni_to_clean_str(file_content))
tokens_new = nltk.pos_tag(tokens)
```
